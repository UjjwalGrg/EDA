```python
df.rename(columns = {"Country Name" : "Country"}, inplace =  True)
df.rename(columns = {"Current Population" : "Population 2023"}, inplace = True)
df.rename(columns = {"Area" : "Area(km²)"}, inplace = True)
df.rename(columns = {"Land Area" : "Land Area(km²)"}, inplace = True)
df.rename(columns = {"Density" : "Density(/km²)"}, inplace = True)
df.rename(columns = {"World Percentage" : "World %"}, inplace = True)
```

```python
df["Population 2023"] = df["Population 2023"].str.replace(',','')
df["Population 2022"] = df["Population 2022"].str.replace(',','')
df["Growth Rate"] = df["Growth Rate"].str.replace('%','')
df["World %"] = df["World %"].str.replace('%','')
df["Density(/km²)"] = df["Density(/km²)"].str.replace(',','')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
m = df.head(10)
m
```

```python
fig, axs = plt.subplots(1,2,figsize= (30,10))

sns.barplot(x = "Country", y = "Growth Rate", data = m,ax = axs[0])
axs[0].set_title("Growth Rate(2022 - 2023)")
axs[0].set_xticklabels(labels = m["Country"],rotation=90)

sns.barplot(x = "Country", y = "Population 2023", data = m,ax = axs[1])
axs[1].set_title("Population 2023")
axs[1].set_xticklabels(labels = m["Country"],rotation=90)

plt.suptitle("Top 10 Highly Populated Countries")
```

```python
plt.figure(figsize = (30,10))
plt.suptitle("Top 10 Highly Populated Countries")

plt.subplot(1,2,1)
sns.barplot(x = "Country", y = "Growth Rate", data = m)
plt.title("Growth Rate 2022-2023")
plt.xticks(rotation=90)

plt.subplot(1,2,2)
sns.barplot(x = "Country", y = "Population 2023", data = m)
plt.title("Population 2023")
plt.xticks(rotation=90)

plt.show()
```

```python
pos_gr = df[df["Growth Rate"] > 0].sort_values("Growth Rate",ignore_index = True, ascending = False)
pos_gr
```

```python
plt.figure(figsize = (20,10))
plt.title("Top 5 Countries with Fastest Population Growth Rate")

sns.lineplot(x = "Country", y = "Growth Rate", data = pos_gr.head(), marker = "*", color = "red", markersize = 25, linewidth = 5)
plt.xticks(rotation=90)

plt.show()
```

```python
neg_gr = df[df["Growth Rate"] < 0].sort_values("Growth Rate",ignore_index = True)
neg_gr
```

```python
plt.figure(figsize = (20,10))
plt.title("Top 5 Countries with Fastest Population Decline Rate")

sns.lineplot(x = "Country", y = "Growth Rate", data = neg_gr.head(), marker = "*", color = "red", markersize = 25, linewidth = 5)
plt.xticks(rotation=90)

plt.show()
```
